```python
# Task 2: Mark the success/failed launches for each site on the map

# Fetch the SpaceX launch data
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Extract relevant columns
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Initialize map (centered at a point, e.g., NASA Johnson Space Center)
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Loop through the DataFrame and add markers based on the launch success or failure
for _, row in spacex_df.iterrows():
    launch_site = row['Launch Site']
    lat = row['Lat']
    lon = row['Long']
    launch_class = row['class']
    
    # Set marker color based on launch success or failure
    if launch_class == 1:  # Successful launch
        marker_color = 'green'
        marker_icon = '✓'  # Success icon
    else:  # Failed launch
        marker_color = 'red'
        marker_icon = '✘'  # Failure icon

    # Add marker for each launch event
    folium.Marker(
        location=[lat, lon],
        popup=f'{launch_site} - {"Success" if launch_class == 1 else "Failure"}',
        icon=folium.Icon(color=marker_color, icon=marker_icon)
    ).add_to(site_map)

# Save the map as an HTML file
site_map.save('launch_success_failure_map.html')

# Optionally, display the map inline (if in a Jupyter notebook or compatible environment)
site_map
```

```python

import folium
from folium.plugins import MarkerCluster
import pandas as pd
from js import fetch
import io

# Fetch the SpaceX launch data
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Extract relevant columns
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Define a function to check the value of 'class' and return corresponding marker color
def get_marker_color(class_value):
    if class_value == 1:
        return 'green'  # Success
    else:
        return 'red'  # Failure

# Apply the function to create a new 'marker_color' column
spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)

# Initialize map centered at a specific location
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Loop through the DataFrame to add markers based on 'marker_color'
for _, row in spacex_df.iterrows():
    launch_site = row['Launch Site']
    lat = row['Lat']
    lon = row['Long']
    launch_class = row['class']
    marker_color = row['marker_color']
    
    # Set marker icon based on success or failure
    marker_icon = '✓' if launch_class == 1 else '✘'  # Success: ✓, Failure: ✘
    
    # Add marker to the cluster with appropriate color and icon
    folium.Marker(
        location=[lat, lon],
        popup=f'{launch_site} - {"Success" if launch_class == 1 else "Failure"}',
        icon=folium.Icon(color=marker_color, icon=marker_icon)
    ).add_to(marker_cluster)

# Save the map as an HTML file
site_map.save('launch_success_failure_cluster_map.html')

# Optionally, display the map inline (if in a Jupyter notebook or compatible environment)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians
```

```python
import folium
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians

# Create a map centered on the launch site
launch_site_coord = [28.5721, -80.6480]  # Example: Kennedy Space Center
site_map = folium.Map(location=launch_site_coord, zoom_start=5)

# Function to calculate distance using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Add MousePosition to the map to show lat, lon
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Add a marker for the launch site
folium.Marker(location=launch_site_coord, popup="Launch Site").add_to(site_map)

# Assuming the closest coastline is at a predefined coordinate (near the launch site)
coastline_coord = [28.56367, -80.57163]  # Example: Coastline near Kennedy Space Center
folium.Marker(location=coastline_coord, popup="Coastline Point", icon=folium.Icon(color='blue')).add_to(site_map)

# Calculate the distance between the launch site and the coastline
distance_km = calculate_distance(launch_site_coord[0], launch_site_coord[1], coastline_coord[0], coastline_coord[1])
print(f"Distance between launch site and coastline point: {distance_km:.2f} km")

# Display the map
site_map
```
